## Our data is ready! Let's now develop a code for Logistic regression using numpy


### Suppose y = θ0 + θ1\*x1 + θ2\*x2 + θ3\*x3 + ...
##### y = Final Result ; x = attributes; θ = weights ; θ0 = intercept

### What is the sigmoid function?
![sigmoid](images/sigmoid.png)

### Updating weights matrix at each iteration by minimizing error (alpha = learning rate)
![theta](images/theta.png)

### What is the loss function?
![sigmoid](images/loss.png)

```python
import numpy as np
class LogisticRegression:
    def __init__(self, lr=0.01, num_iter=100000, fit_intercept=True, verbose=False):
        self.lr = lr
        self.num_iter = num_iter
        self.fit_intercept = fit_intercept
        self.verbose = verbose
    
    def __add_intercept(self, X):
        intercept = np.ones((X.shape[0], 1))
        return np.concatenate((intercept, X), axis=1)
    
    def __sigmoid(self, z):
        return 1 / (1 + np.exp(-z))
    def __loss(self, h, y):
        return (-y * np.log(h) - (1 - y) * np.log(1 - h)).mean()
    
    def fit(self, X, y):
        if self.fit_intercept:
            X = self.__add_intercept(X)
        
        # weights initialization
        self.theta = np.zeros(X.shape[1])
        
        for i in range(self.num_iter):
            z = np.dot(X, self.theta)
            h = self.__sigmoid(z)
            gradient = np.dot(X.T, (h - y)) / y.size
            self.theta -= self.lr * gradient
            
            z = np.dot(X, self.theta)
            h = self.__sigmoid(z)
            loss = self.__loss(h, y)
                
            if(self.verbose ==True and i % 10000 == 0):
                print(f'loss: {loss} \t')
    
    def predict_prob(self, X):
        if self.fit_intercept:
            X = self.__add_intercept(X)
    
        return self.__sigmoid(np.dot(X, self.theta))
    
    def predict(self, X):
        return self.predict_prob(X).round()
```

## The model is ready! Let's split the data into training and testing sets and test our model.


```python
from sklearn.model_selection import train_test_split
X = df.drop('Survived', axis = 1)
y = df['Survived']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.33, random_state = 0)
```

```python
model = LogisticRegression(lr=0.1, num_iter=60000)
```

```python
%time model.fit(X_train, y_train)
```

```python
preds = model.predict(X_test)
```

```python
# Testing accuracy
acc=(preds == y_test).mean()
print(f"Accuracy of the model is {acc*100}%")
```
